```python
###########################
# PORTER, NICOLO RYNE A.
# CSIT327 - G1
# LABORATORY 4
# 9/25/2024
###########################

# Data Structures

## Defining Ticket Class

class TicketInformation:

    ticket_id_counter = 1000

    def __init__(self, customer_name, issue, priority):
        self.ticket_id = TicketInformation.ticket_id_counter
        TicketInformation.ticket_id_counter += 1
        self.customer_name = customer_name
        self.issue = issue
        self.priority = priority
        self.status = "UNRESOLVED"
        self.assigned_to = None
    
    def mark_resolved(self):
        if self.status == "UNRESOLVED" : self.status = "RESOLVED"
    
    def assign_to(self, support_staff):
        self.assigned_to = support_staff

## Instantiating ticket sets
tickets = set()
open_tickets = set()
closed_tickets = set()

## Instantiating ticket dictionary
tickets_information = {}

## Instantiating support staff list
support_staff = ['Scooby Doo', 'Jerry Rogers', 'Bart Simpson', 'Shaggy Duck', 'Peter Griffin']

## Instantiating staff assignment
staff_assignments = {staff: [] for staff in support_staff}

# Functions / Feature Functions

## Create
def create_ticket(customer_name, issue, priority):
    ticket = TicketInformation(customer_name, issue, priority)

    tickets_information[ticket.ticket_id] = ticket
    tickets.add(ticket.ticket_id)
    open_tickets.add(ticket.ticket_id)

## Read
def view_open_tickets():
    print("\nVIEWING OPEN TICKETS")
    print("Ticket ID       Customer Name             Issue                          Status          Assigned To")
    print("=" * 120)
    for ticket_id in open_tickets:
        ticket = tickets_information[ticket_id]
        assigned_staff = ticket.assigned_to if ticket.assigned_to is not None else "NOT ASSIGNED"
        print(f"{ticket.ticket_id:<15} {ticket.customer_name:<25} {ticket.issue:<30} {ticket.status:<15} {assigned_staff:<25}")

## Assign
def assign_ticket(ticket_id, support_staff):
    if ticket_id in open_tickets:
        ticket = tickets_information[ticket_id]
        ticket.assign_to(support_staff)
        staff_assignments[support_staff].append(ticket_id) 
    else:
        print(f"ERROR: Ticket ID {ticket_id} is either resolved or does not exist.")
    
   

## Resolve
def resolve_ticket(ticket_id):
    if ticket_id in open_tickets:
        ticket = tickets_information[ticket_id]
        ticket.mark_resolved()
        open_tickets.remove(ticket_id)
        closed_tickets.add(ticket_id)
        print(f"Ticket {ticket_id} has been resolved.")
    else:
        print(f"ERROR: Ticket ID {ticket_id} is either resolved or does not exist.")

# Instantiate Demo Tickets
def start_demo_tickets():
    
    create_ticket("Alice", "Cannot login to email", "High")
    create_ticket("Bob", "System crash during startup", "Medium")
    create_ticket("Winnie the Pooh", "Not enough honey", "High")

# Main Function

def main():
    choice = 0
    start_demo_tickets()

    while choice != 5:
        print("\n--- Ticket Management System ---")
        print("1. Create a Ticket")
        print("2. View Open Tickets")
        print("3. Assign Ticket")
        print("4. Resolve Ticket")
        print("5. Exit\n")
        
        while True:
            temp_choice = input("Enter your choice (1-5): ")

            if isinstance(temp_choice, str) and temp_choice.isdigit():
                choice = int(temp_choice)
                break
            else:
                print("ERROR: Please input a valid choice.")
        
        if choice == 1:
            customer_name = input("Enter customer name: ")
            issue = input("Enter issue description: ")
            priority = input("Enter priority (High, Medium, Low): ")
            create_ticket(customer_name, issue, priority)
            print("Ticket created successfully.")
            
        
        elif choice == 2:
            view_open_tickets()
           
        
        elif choice == 3:
            ticket_id = int(input("Enter Ticket ID to assign: "))
            staff_member = input("Enter support staff name: ")
            assign_ticket(ticket_id, staff_member)
            print(f"Ticket ID {ticket_id} assigned to {staff_member}.")
           
        
        elif choice == 4:
            ticket_id = int(input("Enter Ticket ID to resolve: "))
            resolve_ticket(ticket_id)
           
        
        elif choice == 5:
            print("Exiting the Ticket Management System.")
            

        

if __name__ == "__main__":
    main()
```
