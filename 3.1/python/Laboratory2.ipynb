## Exercise 1: Write a function to calculate the factorial of a given positive integer.
<b>Instructions</b>
> - Define a function `factorial(n)` that takes a positive integer `n` as input.
> - The function should return the factorial of `n`.
> - Use a loop to calculate the factorial.

```python
def factorial(n):
    fac = 1
    for i in range(1, n + 1):
       fac *= i
       
    return fac
 
 
# if __name__ == "__main__":
#    while True:
#       try:    
#          n = int(input("Input a positive integer: "))
#          if(n < 0):
#             print("Please input a positive integer.")
#          elif(n == 0):
#             print("The factorial of 0 is 1")
#             break
#          else:
#             print(f"The factorial of {n} is {factorial(n)}")
#             break
#       except ValueError:
#          print("Please input a valid integer.")
```

## Exercise 2:  Temperature Converter
Write a Python function that converts between Fahrenheit and Celsius.

### Input values:
> User provides a temperature and selects the conversion type (Fahrenheit and Celsius or vice-versa)

### Output value:
The converted temperature

Example

<code>
Enter Temperature: 32
Select a conversion type (1 - Fahrenheit to Celsius, 2 - Celsius to Fahrenheit): 1

Output value:
Converted Temperature: 0.0°C
</code>


```python
def tempConvert(value, type):
    if(type == 1): # Fahrenheit to Celsius
        value = ((value - 32) * 5) / 9
    else: # Celsius to Fahrenheit
        value = value * (9 / 5) + 32
    
    return value

 
 
# if __name__ == "__main__":
#    while True:
#       try:    
#          n = float(input("Enter temperature: "))
#          type = int(input("Select a conversion type (1 - Fahrenheit to Celsius) (2 - Celsius to Fahrenheit): "))
         
#          if(type > 2):
#              print("Please select a valid conversion type.")
#          else:
#             unit = "C" if type == 1 else "F" 
#             print(f"Converted Temperature: {tempConvert(n, type):.1f}°{unit}")
#             break
#       except ValueError:
#          print("Please input a valid number.")
```

## Exercise 3: BMI Calculator
Create a Body Mass Index Calculatore based on user input. User provides weight (in Kilograms) and height (in meters).

Instructions:
- Create a function that accepts a weight and height and returns the BMI.
- create a function that determines the BMI category

BMI Categories:
  + BMI < 18.5: Underweight
  + BMI >= 18.5 and BMI < 24.9: Normal Weight
  + BMI > 24.9 and BMI < 29.9: Overweight
  + otherwise Obesity

Example:

`Enter weight (in kilograms): 70`

`Enter height (in meters): 1.75`

`Output value:`

`BMI: 22.9`

`Category: Normal Weight`

```python
def get_bmi(weight, height):
   return weight / (height*height)

def determine_bmi_category(bmi):
    if(bmi < 18.5):
       return "Underweight"
    elif(bmi >= 18.5 and bmi < 24.9):
       return "Normal Weight"
    elif(bmi > 24.9 and bmi < 29.9):
       return "Overweight"
    else:
       return "Obesity" 
 
 
# if __name__ == "__main__":
#    while True:
#       try:    
#          weight = float(input("Enter weight (in kilograms): "))
#          height = float(input("Enter height (in meters): "))
         
#          if(weight < 0.0 or height < 0.0):
#             print("Please input a positive number.")
#          else:
#             bmi = get_bmi(weight, height)
#             category = determine_bmi_category(bmi)
#             print("Output value:\n")
#             print(f"BMI: {bmi:.1f}")
#             print(f"Category: {category}")
#             break
#       except ValueError:
#          print("Please input a valid number.")
```

### Exercise 5: Pascal's Triangle
Write a Python function that prints out the first n rows of Pascal's triangle
Note : Pascal's triangle is an arithmetic and geometric figure first imagined by Blaise Pascal

    1
  1 2 1
 1 3 3 1
1 4 6 4 1

Each number is the two numbers above it added together

```python
def get_pascal_triangle(n):
   if n <= 0:
      return
   
   pascal_list = [1]
   
   for i in range(n):
      print(" " * (n - i - 1) + " ".join(map(str, pascal_list))) 
      
      new_row = [1]
      for j in range(1, len(pascal_list)):
         new_row.append(pascal_list[j - 1] + pascal_list[j])
      new_row.append(1)
      
      pascal_list = new_row
 

# if __name__ == "__main__":
#    while True:
#       try:    
#          n = int(input("Enter a number: "))
         
#          if(n < 0):
#              print("Please input a positive number.")
#          else:
#             get_pascal_triangle(n)
#             break
#       except ValueError:
#          print("Please input a valid number.")
```

### Exercise 5: Fibonacci
Write a Python function to find and print the first 10 Fibonacci numbers using a while loop

```python

def fib():
   a, b = 0, 1
   count = 0
   while(count < 10):
      print(a, end=" ")
      a, b = b, a + b
      count += 1

# if __name__ == "__main__":
#    while True:
#       try:    
#          n = int(input("Enter a number: "))
         
#          if(n < 0):
#              print("Please input a positive number.")
#          else:
#             fib(n)
#             break
#       except ValueError:
#          print("Please input a valid number.")

```
